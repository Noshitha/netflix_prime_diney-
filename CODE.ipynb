# VISUALIZATIONS

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,8))
plt.title("Movies count over the years",fontsize=25)
sns.histplot(dt.Year);
```

As expected, the vast majority of the movies are from the last 20 years. However, it is also interesting to see that there is a significant amount of old movies in the catalogs too

```python
Hulu_dt = dt.loc[(dt['Hulu'] > 0) ]
Disney_dt = dt.loc[(dt['Disney+'] > 0) ]
PrimeVideo_dt = dt.loc[(dt['Prime Video'] > 0)  ]
Netflix_dt = dt.loc[(dt['Netflix'] > 0)  ]
```

```python
columns = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']
ott_dt = dt[columns]
#Counting all values for each platform
ott = ott_dt.sum()
#Plotting into a bar plot using Seaborn
sns.set_style('darkgrid')
plt.figure(figsize=(12,8))
plt.title('Available Movies and TV Shows',fontsize=25)
sns.barplot(x=ott.index, y=ott.values);
```

Amazon prime video have majority of content followed by netflix while disney+ have the least content
